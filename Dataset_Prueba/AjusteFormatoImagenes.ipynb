```python
##CON ESTE CÓDIGO CAMBIAMOS LOS NOMBRES DE LAS IMÁGENES, MANTENIENDO SU FORMATO

from PIL import Image
import os
import uuid
import shutil  # Importa el módulo shutil para copiar archivos

def asignar_nombres_unicos(carpeta_entrada, carpeta_salida):
    # Crear la carpeta de salida si no existe
    if not os.path.exists(carpeta_salida):
        os.makedirs(carpeta_salida)

    # Listar todos los archivos en la carpeta de entrada
    archivos_entrada = os.listdir(carpeta_entrada)

    for archivo in archivos_entrada:
        ruta_entrada = os.path.join(carpeta_entrada, archivo)

        # Verificar si el archivo es una imagen
        if os.path.isfile(ruta_entrada) and archivo.lower().endswith(('.webp', '.jfif', '.avif', '.png', '.jpg', '.jpeg', '.gif', '.bmp')):

            # Generar un nombre único alfanumérico para la nueva imagen
            nuevo_nombre = str(uuid.uuid4()) + os.path.splitext(archivo)[-1]

            # Crear la ruta completa para la nueva imagen en la carpeta de salida
            ruta_salida = os.path.join(carpeta_salida, nuevo_nombre)

            # Copiar el archivo a la carpeta de salida con el nuevo nombre
            shutil.copy(ruta_entrada, ruta_salida)

            print(f'Archivo copiado: {ruta_entrada} a {ruta_salida}')

# Especifica la carpeta de entrada y salida
carpeta_entrada = '/content/drive/Shareddrives/TDG 2023/Dataset_Prueba/Multiples ciclistas/Simples'
carpeta_salida = '/content/drive/Shareddrives/TDG 2023/Dataset_Prueba/Multiples ciclistas/SR'

# Realizar la asignación de nombres únicos
asignar_nombres_unicos(carpeta_entrada, carpeta_salida)
```

```python
"""Este código nos sirve para tomar las diferentes imágenes
   del dataset de prueba y dejarlas en el mismo formato: *.jpeg"""

from PIL import Image
import os

def convertir_a_jpeg(archivo_entrada, archivo_salida):
    try:
        # Abrir la imagen original
        imagen = Image.open(archivo_entrada)

        # Guardar la imagen en formato JPEG
        imagen.save(archivo_salida, 'JPEG')

        print(f'Conversión exitosa: {archivo_entrada} a {archivo_salida}')

    except Exception as e:
        print(f'Error al convertir {archivo_entrada}: {str(e)}')

def convertir_grupo_a_jpeg(carpeta_entrada, carpeta_salida):
    # Crear la carpeta de salida si no existe
    if not os.path.exists(carpeta_salida):
        os.makedirs(carpeta_salida)

    # Listar todos los archivos en la carpeta de entrada
    archivos_entrada = os.listdir(carpeta_entrada)

    for archivo in archivos_entrada:
        ruta_entrada = os.path.join(carpeta_entrada, archivo)
        ruta_salida = os.path.join(carpeta_salida, archivo.replace(os.path.splitext(archivo)[-1], '.jpeg'))

        # Convertir solo si es un archivo de imagen
        if os.path.isfile(ruta_entrada) and archivo.lower().endswith(('.webp', '.jfif', '.avif', '.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            convertir_a_jpeg(ruta_entrada, ruta_salida)

# Especifica la carpeta de entrada y salida
carpeta_entrada = '/content/drive/Shareddrives/TDG 2023/Dataset_Prueba/Multiples ciclistas/SR'
carpeta_salida = '/content/drive/Shareddrives/TDG 2023/Dataset_Prueba/Multiples ciclistas/SimplesJPEG'

# Realizar la conversión
convertir_grupo_a_jpeg(carpeta_entrada, carpeta_salida)
```

```python
#ESTE CÓDIGO, CAMBIA LOS NOMBRES DE LAS IMÁGENES JPEG, PARA QUE SIGAN UN CONSECUTIVO

import os

def renombrar_imagenes(carpeta_entrada, nombre_base):
    # Listar todos los archivos en la carpeta de entrada
    archivos_entrada = os.listdir(carpeta_entrada)

    # Filtrar solo archivos de imagen
    imagenes = [archivo for archivo in archivos_entrada if os.path.isfile(os.path.join(carpeta_entrada, archivo)) and archivo.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]

    # Ordenar las imágenes por nombre
    imagenes.sort()

    # Renombrar las imágenes
    for i, imagen in enumerate(imagenes, start=1):
        nuevo_nombre = f'{nombre_base}_{i}{os.path.splitext(imagen)[-1]}'
        ruta_entrada = os.path.join(carpeta_entrada, imagen)
        ruta_salida = os.path.join(carpeta_entrada, nuevo_nombre)

        os.rename(ruta_entrada, ruta_salida)
        print(f'Imagen renombrada: {ruta_entrada} a {ruta_salida}')

# Especifica la carpeta de entrada y el nombre base
carpeta_entrada = '/content/drive/Shareddrives/TDG 2023/Dataset_Prueba/Multiples ciclistas/NocturnosJPEG'
nombre_base = 'NocturnosMultiples'

# Realizar el proceso de renombrar
renombrar_imagenes(carpeta_entrada, nombre_base)
```
